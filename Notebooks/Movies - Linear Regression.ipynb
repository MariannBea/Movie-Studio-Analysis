```python
import pandas as pd
import numpy as np
import math

from pandas_profiling import ProfileReport

import seaborn as sns
import matplotlib.pyplot as plt


from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
recent_profitable_movies = pd.read_csv("recent_profitable_movies.csv")
```

#### Does the number of movies a director or actor was in correlated with increased profits?

To investigate this, I used simple linear regression with each type of person (actor1, actor2, actor3 and director).  I also used multiple linear regression combining all of these variables.

```python
X = recent_profitable_movies[['director_count']]
y = recent_profitable_movies['profit'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)

# evaluate the model
y_pred = model.predict(X_test)
# evaluate predictions
mean_absolute_error = metrics.mean_absolute_error(y_test, y_pred)
squared_error = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
coefficient = model.score(X, y)
slope = model.coef_
print('Mean Absolute Error: %.3f' % mean_absolute_error)
print('Squared Error: %.3f' % squared_error)
print('Coefficient of determination: %.3f' % coefficient)
print('Slope: %.3f' % slope)

plt.scatter(X_test, y_test,  color='gray')
plt.plot(X_test, y_pred, color='c', linewidth=3)
plt.show()
```

#### Are any genres correlated with more profits?

To look at this, I used multiple linear regression with profits as the dependent variable.

```python
X = recent_profitable_movies[['News', 'Thriller', 'Fantasy', 'Comedy',
       'Animation', 'History', 'Western', 'Music', 'Family', 'Short',
       'Game-Show', 'Action', 'Sport', 'Drama', 'Mystery', 'Sci-Fi',
       'Biography', 'Documentary', 'War', 'Film-Noir', 'Horror', 'Crime',
       'Reality-TV', 'Musical', 'Adventure', 'Romance']]
y = recent_profitable_movies['normalized_profit'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

# evaluate the model
y_pred = model.predict(X_test)
# evaluate predictions
# evaluate predictions
mean_absolute_error = metrics.mean_absolute_error(y_test, y_pred)
squared_error = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
coefficient = model.score(X, y)
slope = model.coef_
print('Mean Absolute Error: %.3f' % mean_absolute_error)
print('Squared Error: %.3f' % squared_error)
print('Coefficient of determination: %.3f' % coefficient)

sns.regplot(x=y_test, y=y_pred, ci=None, scatter_kws={"color": "gray"}, line_kws={"color": "red"}, fit_reg = True, robust = True)
```

#### Analysis of Results

With a correlation coefficent of 0.243, it seems there is more of a correlation between genre and number of movies or facebooks likes of actors or directors.
