```python
#used to store and manipulate data
import pandas as pd
import numpy as np
#import math

#files for plotting
import seaborn as sns
import matplotlib.pyplot as plt

# files needed for linear regression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression

# needed to print evaluation metrics
from sklearn import metrics
```

```python
#read in cleaned dataset

movies_df = pd.read_csv('movies_df_cleaned.csv')
```

```python
#Look at the correlation between genres and profits
X = movies_df[['Family', 'Musical', 'Documentary', 'Sport', 'Crime', 'Sci-Fi', 
               'History', 'Adventure', 'Western', 'Music', 'Mystery', 'Horror', 
               'Fantasy', 'Drama', 'News', 'Romance', 'Comedy', 'Biography', 'War', 
               'Action', 'Thriller', 'Animation']]
y = movies_df['normalized_profit'].values.reshape(-1,1)

#divide data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

#create the model and run it with the training sets
model = LinearRegression()
model.fit(X_train, y_train)

# evaluate the model
y_pred = model.predict(X_test)

# evaluate predictions
mean_absolute_error = metrics.mean_absolute_error(y_test, y_pred)
squared_error = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
coefficient = model.score(X, y)

#print the results
print('Mean Absolute Error: %.3f' % mean_absolute_error)
print('Squared Error: %.3f' % squared_error)
print('Coefficient of determination: %.3f' % coefficient)

#plot the results
fig, ax = plt.subplots(figsize = (5, 3))
graph = sns.regplot(x=y_test, y=y_pred, ci=None, scatter_kws={'color': 'gray'}, 
            line_kws={'color': 'red'}, fit_reg = True, robust = True).set_title('All Genres and Profits')
```
