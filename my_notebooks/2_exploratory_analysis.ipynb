```python
genres = ['Family', 'Musical', 'Sport', 'Crime', 
          'Sci-Fi', 'History', 'Adventure', 'Western', 'Music', 
          'Mystery', 'Horror', 'Fantasy', 'Drama', 'Romance', 
          'Comedy', 'Biography', 'War', 'Action', 'Thriller', 
          'Animation', 'Documentary']

flops = movies_df.loc[movies_df['profit_infl'] <= 0].reset_index(drop = True)
hits = movies_df[movies_df['profit_infl'] >= 11929309505.107433].reset_index(drop = True)

flops_hits_list = []

for genre in genres:
    flops_hits_list.append((genre, flops[genre].value_counts()[1], hits[genre].value_counts()[1]))
           
flops_hits = pd.DataFrame(flops_hits_list)
flops_hits.columns = 'genre', 'flops', 'hits'

x = np.arange(len(flops_hits['genre']))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize = (10, 6))
hits = ax.bar(x - width/2, flops_hits['hits'], width, label='Hits', color = '#009292')
flops = ax.bar(x + width/2, flops_hits['flops'], width, label='Flops', color = '#920000')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Movies')
ax.set_title('Flops and Hits By Genre')
ax.set_xticks(x)
ax.set_xticklabels(flops_hits['genre'], rotation = 45)
ax.legend()

fig.tight_layout()

plt.show()
```

```python
# create data frams with movies that lost money, were in the top 25th percentile and movies between these two points
flops = movies_df.loc[movies_df['profit_infl'] <= 0].reset_index(drop = True)
hits = movies_df[movies_df['profit_infl'] >= 11929309505.107433].reset_index(drop = True)
average = movies_df[movies_df['profit_infl'].between(0, 11929309505.107433)].reset_index(drop = True)

#determing value counts by rating
flops_df = pd.DataFrame(flops.content_rating.value_counts().reset_index().values, columns=['rating', 'flops'])
hits_df = pd.DataFrame(hits.content_rating.value_counts().reset_index().values, columns=['rating', 'hits'])
average_df = pd.DataFrame(average.content_rating.value_counts().reset_index().values, columns=['rating', 'average'])

#merge all of the data frames togehter
flops_hits = pd.merge(flops_df, hits_df, on = 'rating')
flops_hits = pd.merge(flops_hits, average_df, on = 'rating')

x = np.arange(len(flops_hits['rating']))  # the label locations
width = 0.25  # the width of the bars

#create a bar graph
fig, ax = plt.subplots(figsize = (10, 6))
hits = ax.bar(x + (width + .01), flops_hits['hits'], width, label='Hits', color = '#009292')
flops = ax.bar(x - (width + .01), flops_hits['flops'], width, label='Flops', color = '#920000')
average = ax.bar(x, flops_hits['average'], width, label='Average', color = '#000000')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Movies')
ax.set_title('Flops and Hits By Content Rating')
ax.set_xticks(x)
ax.set_xticklabels(flops_hits['rating'], rotation = 45)
ax.legend()

fig.tight_layout()

plt.show()
```
